```python
import matplotlib.pyplot as plt
import seaborn as sns

# Datatentang jenis produk dan penjualan
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [230, 400, 350, 280]

# bar plot yang menarik untuk membuat audiens lebih tertarik
plt.figure(figsize=(10, 6))
sns.barplot(x=products, y=sales, palette='viridis')

plt.title('Sales Comparison of Different Products (Customer Exhibition)', fontsize=16, weight='bold')
plt.xlabel('Product', fontsize=12)
plt.ylabel('Sales (in Units)', fontsize=12)
plt.grid(True, axis='y', linestyle='--', alpha=0.5)

for i, v in enumerate(sales):
    plt.text(i, v + 10, str(v), ha='center', fontsize=12)

# Menampilkan plot
plt.tight_layout()
plt.show()
```



*  Tone: Informatif, visual, mudah dimengerti oleh audiens yang luas.
*   Tujuan: Menunjukkan data yang menarik secara visual untuk menjangkau banyak pelanggan.
*   Analisis: Fungsi exhibition, karena tujuan utamanya adalah pameran data dalam cara yang menarik perhatian dan mudah dipahami oleh setiap kalangan audiens.





Tujuan dari visualisasi data diatas adalah untuk memamerkan perbandingan penjualan produk secara visual kepada audiens atau calon customer dalam rangka promosi dengan fungsi exhibition visualisasi in digunakan untuk menampilkan data yang menarik kepada audiens yang lebih besar dan non spesialis, sehingga lebi mudah dimengerti tiap kalangan audiens.
